```python
#APPROACH1

def match_prefix(s,l):
    results = set()
    l = set(l)
    for word in l:
        if word.startswith(s):
            results.add(word)
    return results

s = 'de'
l = ['dog', 'deer', 'deal']
match_prefix(s,l)
```

```python
class Trie():
    
    def __init__(self):
        self._trie = {} 
    
    def insert(self,word):
        trie = self._trie
        for char in word:
            if char not in trie:
                trie[char] = {}
            trie = trie[char]
        trie['*'] = True
    
    def elements(self,prefix):
        trie = self._trie
        for char in prefix:
            if char in trie:
                trie = trie[char]
            else:
                return []
        return self._element(trie)
    
    def _element(self,b):
        results = []
        for s,v in b.items():
            if s == '*':
                subresult = ['']
            else:
                subresult = [s + k for k in self._element(v)]
            results.append(subresult[0])
        return results
            
    

```

```python
s = 'de'
words = ['dog', 'deer', 'deal']
trie = Trie()
for word in words:
    trie.insert(word)

def match_prefix(s):
    suffixes = trie.elements(s)
    return [s + w for w in suffixes]
match_prefix(s)
```

Time Complexity: O(N).
