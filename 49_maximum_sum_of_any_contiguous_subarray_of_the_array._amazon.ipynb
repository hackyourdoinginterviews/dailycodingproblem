```python
def max_contiguous_subarray_sum(nums):
    max_so_far,max_ending_here=0,0
    for n in nums:
        max_ending_here=max_ending_here + n
        if max_ending_here < 0:
            max_ending_here=0
        if max_so_far < max_ending_here:
            max_so_far=max_ending_here
    return max_so_far
```

```python
nums=[34, -50, 42, 14, -5, 86] 
print(max_contiguous_subarray_sum(nums))
```

```python
nums=[-5, -1, -8, -9]
print(max_contiguous_subarray_sum(nums))
```

```python
def max_contiguou_subarray_sum_dp(nums):
    max_so_far,current_max=nums[0],nums[0]
    for i in range(1,len(nums)):
        current_max=max(nums[i],(current_max+nums[i]))
        max_so_far=max(max_so_far,current_max)
    return max_so_far
```

```python
nums=[34, -50, 42, 14, -5, 86] 
print(max_contiguou_subarray_sum_dp(nums))
```

```python
nums=[-5, -1, -8, -9]
print(max_contiguou_subarray_sum_dp(nums))
```
