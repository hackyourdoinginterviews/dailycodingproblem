```python
class Node(object):
    def __init__(self,data):
        self.data=data
        self.nextNode=None
        
class LinkList(object):
    def __init__(self):
        self.head=None
        self.size=0
     
    #O(1)
    def insert(self,data):
        self.size=self.size+1
        newNode=Node(data)
        print(newNode.data)
        if not self.head:
            self.head=newNode
        else:
            newNode.nextNode=self.head
            self.head= newNode
    
    def sizeoflinkedlist(self):
        return self.size
        
    #O(n)
    def display(self):
        actualNode=self.head
        while actualNode is not None:
            print(actualNode.data)
            actualNode=actualNode.nextNode
    
    def gethead(self):
        actualNode=self.head
        return actualNode
    
```

```python
A= LinkList()
A.insert(10)
A.insert(8)
A.insert(7)
A.insert(3)
```

```python
A.display()
```

```python
B= LinkList()
B.insert(10)
B.insert(8)
B.insert(1)
B.insert(99)
```

```python
B.display()
```

```python
def intersection_of_list(headA,headB):
    if not headA or not headB : return None
    a,b =headA,headB
    while a.data != b.data:
        if not a: 
            a= b.head
        else: 
            a=a.nextNode
        if not b : 
            b=a.head
        else: 
            b=b.nextNode
    return b.data
```

```python
head_of_A=A.gethead()
head_of_B=B.gethead()        
print(intersection_of_list(head_of_A,head_of_B))
```
