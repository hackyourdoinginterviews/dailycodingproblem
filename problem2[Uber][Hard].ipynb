```python
def productExceptSelf(inputnums):
    len_of_nums=len(inputnums)
    print(len_of_nums)
    left, right, result = [0]*len_of_nums, [0]*len_of_nums, [0]*len_of_nums
    #left product computation 
    left[0]=1
    for i in range(1,len_of_nums):
        left[i]=inputnums[i-1]*left[i-1]
    #right product computtaion 
    print("Left Product")
    print(left)
    right[len_of_nums-1]=1
    for i in reversed(range(len_of_nums-1)):
        right[i]=inputnums[i+1]*right[i+1]
    print("Right Product")
    print(right)
    #left * right 
    for i in range(0,len_of_nums):
        result[i]=left[i]*right[i]
    return result 
```

```python
t=productExceptSelf([1,2,3,4,5])
print("Combined left and right product ")
print(t)
```

Time Complexity is => 3 for loops for to compute left array, right array and result => O(n)


Space complexity : O(N) used up by the two intermediate arrays that we constructed to keep track of product of elements to the left and right. 



This is good enough in most interviews but some may have a followup qiestion can you do it in O(1) space?

```python
def productExceptSelf_bigO_1(inputnums):
    len_of_nums=len(inputnums)
    result = [0]*len_of_nums
    result[0]=1
    for i in range(1,len_of_nums):
        result[i]=inputnums[i-1]*result[i-1]
    print("Left Product")
    print(result)
    R=1
    for i in reversed(range(len_of_nums)):
        result[i]=result[i]*R
        R*=inputnums[i]
    return result
```

```python
t=productExceptSelf_bigO_1([1,2,3,4,5])
print("Combined left and right product ")
print(t)
```
