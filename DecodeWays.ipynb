```python
def waysofdecode(data,k,tracker):
    #empty string data=""
    if k==0:
        return 1
    s=len(data)-k
    # starts with zero 
    if data[s]=='0':
        return 0
    #memorization 
    if tracker[k] != "null":
        return tracker[k]
    # waysofdecode(data,k,tracker) = waysofdecode(data,k-1,tracker)
    result=waysofdecode(data,k-1,tracker)
    if k >=2 and int(data[s:s+2])<=26:
        #waysofdecode(data,k,tracker) = waysofdecode(data,k -1 ,tracker) + waysofdecode(data,k -2 ,tracker)
        result=result+waysofdecode(data,k-2,tracker)
    #append data to result array for reuse 
    tracker[k]=result
    return result 
        
    
def numDecodings(s):
    k=len(s)
    #list used to store intermediate results for reuse to redue O (2^n) to O(n) to reduce forking potemntial 
    tracker = ["null"] * (k + 1 )
    return waysofdecode(s,k,tracker)
  
    
```

```python
s="11" 
print(numDecodings(s))
```

Time Complexity: O(N).
