```python
def water_capacity_rain_water(LIST):
    l,r=0,len(LIST)-1
    capacity=0
    lmax,rmax=0,0
    while l<r:
        if LIST[l]<LIST[r]:
            if LIST[l]>lmax:
                lmax=LIST[l]
            else:
                capacity+=lmax-LIST[l]
            l=l+1
        else:
            if LIST[r]>rmax:
                rmax=LIST[r]
            else:
                 capacity+=rmax-LIST[r]
            r=r-1
    return capacity
  
    
```

```python
LIST=[3,0,1,3,0,5]
print(water_capacity_rain_water(LIST))
```

```python
LIST=[0,1,0,2,1,0,1,3,2,1,2,1]
print(water_capacity_rain_water(LIST))
```

Time Complexity: O(N).
