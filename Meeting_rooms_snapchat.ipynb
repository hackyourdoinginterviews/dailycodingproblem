```python
import heapq
def minMeetingRooms(meetings):
    meetings.sort()
    mx = 0
    hp = []
    for start, end in meetings:
        while hp and hp[0] <= start:
            heapq.heappop(hp)
        heapq.heappush(hp, end)
        mx = max(mx, len(hp))
        
    return mx
```

```python
minMeetingRooms([(30, 75), (0, 50), (60, 150)])
```

```python
minMeetingRooms([[7,10],[2,4]])
```

```python
minMeetingRooms([(0,30),(15,20) , (5,10)])
```

Time Complexity: O(NlogN).

space Complexity: O(N)
