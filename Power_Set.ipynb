```python
def pow_set(nums):
    """
    :type nums: List[int]
    :rtype: List[List[int]]
     """
    result = [[]]
    for elem in nums:
        result.extend([x + [elem] for x in result])
    return result
```

```python
nums=[1,2,3]
print(pow_set(nums))
```

Extend has time complexity of O(k). Where k is the length of list which need to be added.

# Append v/s Extend  in case of lists ? 

```python
# Append: Adds its argument as a single element to the end of a list. The length of the list increases by one. 
my_list = ['coding', 'interview'] 
my_list.append('hacks') 
print(my_list) 
```

```python
# extend(): Iterates over its argument and adding each element to the list and extending the list. 
#The length of the list increases by number of elements in it’s argument.
my_list = ['coding', 'interview'] 
new_list= ["leet_code","daily_coding_problems"]
my_list.extend(new_list) 
print(my_list) 
```

Time Complexity:

Append has constant time complexity i.e.,O(1).

Extend has time complexity of O(k). Where k is the length of list which need to be added.

```python
result = [[]]
for elem in [1,2,3]:
    result.extend([x + [elem] for x in result])
    print(result)
print(result)
```
