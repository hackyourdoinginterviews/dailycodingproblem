```python
def singleNumber(nums):
    a,b=0,0
    for n in nums:
        b=(b^n)& (~a)
        a=(a^n)& (~b)
    return b 
```

```python
nums=[0,1,0,1,0,1,99]
singleNumber(nums)
```

```python
nums=[2,2,3,2]
singleNumber(nums)
```
