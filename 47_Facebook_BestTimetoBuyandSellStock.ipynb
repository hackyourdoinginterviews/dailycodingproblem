```python
  def maxProfit(prices):
            if len(prices)==0: return 0
            max_to_here=0
            max_so_far=0
            for i in range(1,len(prices)):
                max_to_here=max(0,max_to_here+prices[i]-prices[i-1])
                max_so_far=max(max_to_here,max_so_far)
            return max_so_far
```

```python
prices=[9, 11, 8, 5, 7, 10]
print(maxProfit(prices))
```
