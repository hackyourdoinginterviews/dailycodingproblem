```python
def find_max_sum(input_arr): 
    include = 0
    exclude = 0
     
    for i in range(len(input_arr)): 
        
        if exclude > include :
            new_exclude = exclude
        else:
            new_exclude = include
          
        include = exclude + input_arr[i] 
        exclude = new_exclude 
      
    # Max of include & exclud eis returned 
    if exclude > include:
        return exclude 
    else:
        return include 
#credits https://www.geeksforgeeks.org/maximum-sum-such-that-no-two-elements-are-adjacent/     
```

```python
input_arr=[2,4,6,2,5]
print(find_max_sum(input_arr))
```
